```python
import ast
from ast import literal_eval
def convert(obj):
  list = []
  for i in ast.literal_eval(obj):
    list.append(i['name'])
  return list
```

```python
convert(df.iloc[0]['genres'])
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df['genres']
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
def convert3(obj):
  list = []
  cnt = 0
  for i in ast.literal_eval(obj):
    if cnt != 3:
      list.append(i['name'])
      cnt += 1
  return list
```

```python
convert3(df['cast'][0])
```

```python
df['cast'].apply(convert3)
```

```python
df['cast'] = df['cast'].apply(convert3)
```

```python
def convertcrew(obj):
  list = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      list.append(i['name'])
  return list
```

```python
convertcrew(df['crew'][0])
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_func(text):
  list = []
  for i in text.split():
    list.append(ps.stem(i))
  return " ".join(list)
```

```python
stem_func('loved is, potted heelo-hello')
```

```python
stem_func(new_df['tag'][0])
```

```python
new_df['tag'] = new_df['tag'].apply(stem_func)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.fit_transform(new_df['tag']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
new_df[new_df['title'] == 'Avatar'].index[0]
```

```python
new_df[new_df['title'] == 'Batman']
```

```python
def recommender(movie):
  movie_idx = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_idx]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommender('Avatar')
```

```python
recommender('Batman')
```
